```python
meteo=pd.read_csv('IGENVEVE2_from_2019-01-01_to_2019-12-31.csv',index_col='Date_Time',parse_dates=[0,])
meteo.head()
```

```python
meteo=meteo.resample('H').mean().interpolate()
meteo.head()
```
